### Question 8
Create a tuple with repeated values and count how many times an element appears.

```python
# write your answer here

t=("apple", "banana", "cherry", "apple", "cherry")

s=[]  #apple , banana #cherry


for i in t:
  if i not in s:
    print(f"this is count of {i}:",t.count(i))
    s.append(i)


```
